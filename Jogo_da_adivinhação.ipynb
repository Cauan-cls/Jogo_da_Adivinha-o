```python
import random
from abc import ABC, abstractmethod

# ABSTRAÇÃO
# Classe abstrata base do jogo

class Jogo(ABC):
    @abstractmethod
    def iniciar(self):
        pass
    @abstractmethod
    def jogar(self):
        pass


# HERANÇA
# Classe que herda da classe Jogo

class JogoAdivinhacao(Jogo):
    def __init__(self, nome_jogador="Jogador"): # Melhoria 1: Adiciona o nome do jogador
        # ENCAPSULAMENTO
        # Atributos privados
        self.__nome_jogador = nome_jogador
        self.__numero_secreto = random.randint(1, 100)
        self.__tentativas = 0
        self.__limite = 10

    # método para iniciar o jogo
    def iniciar(self):
        print(f"\n🎮 Olá, {self.__nome_jogador}! Bem-vindo(a) ao JOGO DA ADIVINHAÇÃO")
        print("Tente adivinhar o número entre 1 e 100")
        print("Você tem", self.__limite, "tentativas")

    # POLIMORFISMO
    # implementação do método abstrato
    def jogar(self):
        while self.__tentativas < self.__limite:
            try:
                palpite = int(input("Digite seu palpite: "))
            except ValueError:
                print("Por favor, digite um número válido.")
                continue

            self.__tentativas += 1
            if palpite == self.__numero_secreto:
                print(f"🎉 Parabéns, {self.__nome_jogador}! Você acertou em {self.__tentativas} tentativas!")
                return self.__tentativas
            elif palpite < self.__numero_secreto:
                print("⬆️ O número secreto é MAIOR")
            else:
                print("⬇️ O número secreto é MENOR")
        print("❌ Você perdeu! Suas tentativas acabaram.")
        print("O número era:", self.__numero_secreto)
        return None


def executar_jogo(jogo: Jogo):
    jogo.iniciar()
    return jogo.jogar()


# PROGRAMA PRINCIPAL

def main():
    nome = input("Por favor, digite seu nome: ") # Melhoria 1: Pede o nome do jogador
    melhor_pontuacao = float('inf') # Melhoria 3: Armazena a melhor pontuação (menor tentativas)
    historico_jogos = [] # Melhoria 3: Armazena histórico de jogos da sessão

    while True: # Melhoria 2: Loop para jogar novamente
        jogo_atual = JogoAdivinhacao(nome)
        tentativas_rodada = executar_jogo(jogo_atual)

        if tentativas_rodada is not None:
            historico_jogos.append({'nome': nome, 'tentativas': tentativas_rodada})
            if tentativas_rodada < melhor_pontuacao:
                melhor_pontuacao = tentativas_rodada
                print(f"🏆 Nova melhor pontuação pessoal: {melhor_pontuacao} tentativas!")

        # Melhoria 3: Exibe o ranking da sessão
        if historico_jogos:
            print("\n--- Ranking da Sessão ---")

            ranking_ordenado = sorted(historico_jogos, key=lambda x: x['tentativas'])
            for i, recorde in enumerate(ranking_ordenado):
                print(f"{i+1}. {recorde['nome']}: {recorde['tentativas']} tentativas")
            print("------------------------")

        jogar_novamente = input("Deseja jogar novamente? (s/n): ").lower()
        if jogar_novamente != 's':
            print("Obrigado por jogar!")
            break

# Executa o programa principal
if __name__ == "__main__":
    main()
```
